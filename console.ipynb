### 4. Trabajando con datos faltantes


```python
# Rellenando datos faltantes
# Se hara una mascara booleana
missing_titles_mask = eluniversal['title'].isna()
missing_titles_mask.head(10)
```

```python
missing_titles = (eluniversal[missing_titles_mask]['url'].
                     str.extract(r'(?P<missing_titles>[^/]+)$')
                     .applymap(lambda title: title.split('-'))
                     .applymap(lambda title_word_list: ' '.join(title_word_list))
                 )
missing_titles
```

```python
eluniversal.loc[missing_titles_mask, 'title'] = missing_titles.loc[:, 'missing_titles']
```

## Limpieza Adicional

### 5. Añadir uid a las filas

```python
import hashlib

uids = (eluniversal
            .apply(lambda row: hashlib.md5(bytes(row['url'].encode())), axis=1)
            .apply(lambda hash_object: hash_object.hexdigest())
       )
eluniversal['uid'] = uids
eluniversal.set_index('uid', inplace = True)

eluniversal
```

### 6. Eliminar saltos de linea en el body

```python
#
stripped_body = (eluniversal
                     .apply(lambda row: row['body'], axis=1)
                     .apply(lambda body: list(body))
                     .apply(lambda letter: list(map(lambda letter: letter.replace('\n', ' '), letter)))
                     .apply(lambda letters: ''.join(letters))
                )
eluniversal['body'] = stripped_body
```

Automatizando en el receipe

### 7. Enriquecer el dataset
Se hara uso de la libreria nltk, la cual procesa lenguaje natural, para contar el número de palabras

```python
%pip install nltk
```

```python
# 6 tokenizar el titulo y el body
import nltk
from nltk.corpus import stopwords

nltk.download('punkt')
nltk.download('stopwords')

stop_words = set(stopwords.words('spanish'))
```

```python
def tokenize_column(df, column_name):
    return (df
               .dropna()
               .apply(lambda row: nltk.word_tokenize(row[column_name]), axis=1)
               .apply(lambda tokens: list(filter(lambda token: token.isalpha(), tokens)))
               .apply(lambda tokens: list(map(lambda token: token.lower(), tokens)))
               .apply(lambda word_list: list(filter(lambda word: word not in stop_words, word_list)))
               .apply(lambda valid_word_list: len(valid_word_list))
           )

eluniversal['n_tokens_title'] = tokenize_column(eluniversal, 'title')
eluniversal['n_tokens_body'] = tokenize_column(eluniversal, 'body')

eluniversal
```
